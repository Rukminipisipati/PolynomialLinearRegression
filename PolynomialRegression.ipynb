```python
#Polynomial Linear Regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('C:/Users/rukmi/OneDrive/Documents/ML Course/Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values #all columns except the last one
y=dataset.iloc[:,-1].values #only the last column
```

```python
#Training linear regression on the entire dataset
from sklearn.linear_model import LinearRegression
regressor_linear=LinearRegression()
regressor_linear.fit(X,y)
```

```python
#Training polynomial regression model on the entire dataset
#This is polynomial.So, we have to create matrix of polynomial features x1,x1^2,x1^3.........x1^n.
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4) #Here we are creating polynomial matrix till degree 2 i.e. x1^2.
X_poly=poly_reg.fit_transform(X)
regressor=LinearRegression()
regressor.fit(X_poly,y)
```

```python
#Visualizing linear Regression Result
plt.scatter(X,y,color="red") # X vs Real y
plt.plot(X,regressor_linear.predict(X),color="blue") # X vs Predicted y for X_poly.
plt.title('Real Vs Predicted (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualizing Polynomial Regression Result
plt.scatter(X,y,color="red") # X vs Real y
plt.plot(X,regressor.predict(X_poly),color="blue") # X vs Predicted y.
plt.title('Real Vs Predicted (Polynomial Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Predicting a new result i.e the salary for the Position level 6.5
#Using Linear regression
regressor_linear.predict([[6.5]])
```

```python
#Using Polynomial Linear Regression
regressor.predict(poly_reg.fit_transform([[6.5]]))
```
